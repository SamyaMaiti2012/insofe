```python
import numpy as np
np.random.seed(123) # Setting random seed
import matplotlib.pyplot as plt
%matplotlib inline
```

####  Defining a function for calculation *pi*

```python
def calc_pi (simulation):
    # Generate random numbers from uniform distribution
    x=np.random.uniform(low=0,high=5,size=simulation) # Generate  uniform random numbers in the range 0,5 for x co- ordinate
    y=np.random.uniform(low=0,high=5,size=simulation) # Generate  uniform random numbers in the range 0,5 for y co- ordinate
    z = np.sqrt(x**2+y**2) # Distance of the point from origin.
    pi_value = 4*(np.sum(z <=5)/len(z)) # 4 * (Q / R^2)
     
    col = []
    for i in z:
        if i<=5:
            col.append('green') # if the point falls in the circle 
        else:
            col.append('brown') # if the point falls in the square
    ## Plot a figure 
    plt.figure(figsize=(10,10))
    plt.scatter(x,y,marker = '*',color = col)
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.title("Monte Carlo simulation for {} times and the value if pi is {}".format(simulation,pi_value),fontsize = 15)
    plt.xlim(0,5)
    plt.ylim(0,5)
    plt.show()

    return ("The value of pi from {} simulations is: {}".format(simulation,pi_value))
    
```

####  Number of simulations

```python
# simulations = [10,100,1000,10000,100000,1000000,10000000]
simulations = [10,100,1000,10000,100000,1000000]
```

```python
for i in simulations:
    print(calc_pi(i))
```
